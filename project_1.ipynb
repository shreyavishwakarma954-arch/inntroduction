```python
df= pd.read_csv('STUDENT.csv')
df
```

```python
print(f"Original 'grade' dtype: {df['grade'].dtype}")
print("Before optimization memory usage:")
print(df['grade'].memory_usage(deep=True))
df['grade']=df['grade'].astype('category')
print(f"Optimization 'grade' dtype:{df['grade'].dtype}")
print("Memory usage after optimization:")
print(df['grade'].memory_usage(deep=True))
```

```python
weights = {
    'math_original':0.4,
    'science_original':0.3,
    'english_original':0.3
}
df['custom_weighted_average']=(
    df['math_original']*weights['math_original']+
    df['science_original']*weights['science_original']+
    df['english_original']*weights['english_original']
)
df['custom_weighted average']=df[list(weights)].mul(list(weights.values()),axis=1).sum(axis=1)
print(df['custom_weighted_average'])
```

```python
#compare attendence between diff student groups
math_scores_by_grade=df.groupby('grade')['math_original'].mean()
print("average math scores by grades:")
print(math_scores_by_grade)
```

```python
attendance_by_grade = df.groupby('grade')['attendance'].mean()
print("average attendence by grades:")
print(attendance_by_grade)
```

```python
#%pip install matplotlib
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.style.use('dark_background')
plt.hist(df['science_original'],bins=10,edgecolor='brown')
plt.title('Distribution of science scores')
plt.xlabel('science score')
plt.ylabel('Number of students')
plt.grid(axis='y',alpha=0.75)
plt.show()
```

```python
#find top performers in english_original test
top_performers=df.nlargest(5,'english_original')
print("top 5 performers in english test:")
print(top_performers[['student_id','english_original']])

```

```python
#apply bonus points to math_original scores
bonus_points=5
df['math_with_bonus'] = df['math_original'] + bonus_points
print("first 5 math scores with bonus:")
print(df[['math_original','math_with_bonus']].head())
```

```python
#scale attendance data based and term based
term_weight=0.8
df['scaled_attendance']=df['attendance'] = term_weight
print("first 5 scaled attendance scored:")
print(df[['attendance','scaled_attendance']].head())
```

```python
#nomalising graddes using broadcasting
min_score=df['science_original'].min()
max_score=df['science_original'].max()
df['science_normalized']=(df['science_original']-min_score)/(max_score-min_score)
print("first 5 normalized science scores:")
print(df[['science_original','science_normalized']].head())
```

```python
#vectorize grade calculation from final score
def assign_grade(score):
    if score >=90: return 'A'
    elif score >=80:return 'B'
    elif score >=70: return 'C'
    elif score >=60:return 'D'
    else:return 'F'
df['calculated_grade']=df['final_score'].apply(assign_grade)
print("first 5 calculated grades:")
print(df[['final_score','calculated_grade']].head())
```

```python
#find attendence eligibility using vectorized comparision
df["eligible"] = df["attendance"] >=75
print(df)
```
